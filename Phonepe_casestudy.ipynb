TASK1.4

```python

num_states = df1['State'].nunique()
num_districts = df4['District'].nunique()
districts_per_state = df4.groupby('State')['District'].nunique()
state_most_districts = districts_per_state.idxmax()
most_districts_count = districts_per_state.max()

print(f"Total number of states: {num_states}")
print(f"Total number of districts: {num_districts}")
print(f"State with the highest number of districts: {state_most_districts} ({most_districts_count})")
```

```python
#TASK 2.1
#Group by state and year to get total transactions and amount
state_yearly_txn = df1.groupby(['State', 'Year']).agg({
    'Transactions': 'sum',
    'Amount (INR)': 'sum'
}).reset_index()

state_yearly_txn
# Identify top 5 states by total transaction volume (across all years)
top_states = state_yearly_txn.groupby('State')['Transactions'].sum().sort_values(ascending=False).head(5).index.tolist()

#Identify bottom 5 states by total transaction volume
bottom_states = state_yearly_txn.groupby('State')['Transactions'].sum().sort_values(ascending=True).head(5).index.tolist()

# Display results
print("Top 5 States by Transaction Volume:")
print(state_yearly_txn[state_yearly_txn['State'].isin(top_states)])

print("\nBottom 5 States by Transaction Volume:")
print(state_yearly_txn[state_yearly_txn['State'].isin(bottom_states)])
```

```python
#TASK 2.2
# Group by state, year, quarter, and transaction type to get total transactions
grouped = df2.groupby(['State', 'Year', 'Quarter', 'Transaction Type'])['Transactions'].sum().reset_index()

# For each state-year-quarter, get the transaction type with max transactions
most_common_types = grouped.sort_values('Transactions', ascending=False).drop_duplicates(subset=['State', 'Year', 'Quarter'])

# Display the results in a tabular format
print("Most Common Transaction Type per State and Quarter:")
print(most_common_types.sort_values(by=['State', 'Year', 'Quarter']))
```

```python
#2.3
# Group by state and brand to sum registered users
brand_usage = df3.groupby(['State', 'Brand'])['Registered Users'].sum().reset_index()

# For each state, find the brand with the highest number of registered users
top_brand_per_state = brand_usage.sort_values('Registered Users', ascending=False).drop_duplicates('State')

# Display result
print("Most Used Device Brand per State:")
print(top_brand_per_state.sort_values('State'))
```

```python
#2.4
# Remove rows with missing or zero population
district_demographics_clean = df5[df5['Population'] > 0]

# Get the district with the highest population in each state
top_districts_by_population = district_demographics_clean.loc[
    district_demographics_clean.groupby('State')['Population'].idxmax()
]

# Display the results
print("Top District by Population in Each State:")
print(top_districts_by_population[['State', 'District', 'Population']].sort_values('State'))
```

```python
# Task 2.5: Calculate Average Transaction Value (ATV) per state

# Group by state to calculate overall average transaction value (weighted)
state_agg = df1.groupby('State').agg({
    'Amount (INR)': 'sum',
    'Transactions': 'sum'
}).reset_index()

# Compute average transaction value
state_agg['ATV'] = state_agg['Amount (INR)'] / state_agg['Transactions']

# Top 5 states by ATV
top_5_atv = state_agg.sort_values('ATV', ascending=False).head(5)

# Bottom 5 states by ATV
bottom_5_atv = state_agg.sort_values('ATV', ascending=True).head(5)

# Display results
print("Top 5 States by Average Transaction Value:")
print(top_5_atv[['State', 'ATV']])

print("\nBottom 5 States by Average Transaction Value:")
print(bottom_5_atv[['State', 'ATV']])
```

```python
# Task 2.6: Analyze app usage trends

# Total number of app opens by year and quarter
app_opens_over_time = df1.groupby(['Year', 'Quarter'])['App Opens'].sum().reset_index()

# Display the aggregated data
print("Total App Opens by Year and Quarter:")
print(app_opens_over_time.sort_values(['Year', 'Quarter']))

# OPTIONAL: Line plot for a selected state (e.g., Maharashtra)
import matplotlib.pyplot as plt

selected_state = 'Maharashtra'
state_data = df1[df1['State'] == selected_state]
state_data['Time'] = state_data['Year'].astype(str) + ' Q' + state_data['Quarter'].astype(str)

plt.figure(figsize=(12, 6))
plt.plot(state_data['Time'], state_data['App Opens'], marker='o', color='teal')
plt.title(f'App Opens Over Time - {selected_state}')
plt.xlabel('Time')
plt.ylabel('App Opens')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Task 2.7: Distribution of transaction types (latest quarter)

# Identify the latest year and quarter
latest_year = df1['Year'].max()
latest_quarter = df1[df1['Year'] == latest_year]['Quarter'].max()

# Filter data for latest year and quarter
latest_data = df2[(df2['Year'] == latest_year) &
                              (df2['Quarter'] == latest_quarter)]

# Group by state and transaction type to sum transactions
txn_distribution = latest_data.groupby(['State', 'Transaction Type'])['Transactions'].sum().unstack(fill_value=0)

# Plot bar chart for a sample state (e.g., Karnataka)
sample_state = 'Karnataka'
txn_distribution.loc[sample_state].plot(kind='bar', figsize=(10, 6), color='skyblue', edgecolor='black')
plt.title(f'Transaction Type Distribution - {sample_state} (Q{latest_quarter}, {latest_year})')
plt.ylabel('Number of Transactions')
plt.xlabel('Transaction Type')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# Task 2.8: Unique mapping of district name and code

# Load required columns
district_mapping = df4[['District', 'Code']].drop_duplicates()

# Check if each district maps to exactly one code
duplicates = district_mapping[district_mapping.duplicated('District', keep=False)]

if duplicates.empty:
    print(" Each district has a unique code.")
else:
    print("The following districts have multiple codes:")
    print(duplicates.sort_values('District'))

# Export unique mapping to CSV
district_mapping.to_csv("district_code_mapping.csv", index=False)
print("\nDistrict-code mapping exported to 'district_code_mapping.csv'")
```

```python
# Task 3.1: Ensure data consistency (State vs District level)

# Aggregate district-level data by state
district_agg = df4.groupby(['State', 'Year', 'Quarter']).agg({
    'Transactions': 'sum',
    'Amount (INR)': 'sum',
    'Registered Users': 'sum',
    'App Opens': 'sum'
}).reset_index()

# Aggregate state-level data
state_agg = df1.groupby(['State', 'Year', 'Quarter']).agg({
    'Transactions': 'sum',
    'Amount (INR)': 'sum',
    'Registered Users': 'sum',
    'App Opens': 'sum'
}).reset_index()

# Merge both summaries on State, Year, Quarter
merged = pd.merge(state_agg, district_agg, on=['State', 'Year', 'Quarter'], suffixes=('_State', '_District'))

# Check for discrepancies
discrepancies = merged[
    (merged['Transactions_State'] != merged['Transactions_District']) |
    (merged['Amount (INR)_State'] != merged['Amount (INR)_District']) |
    (merged['Registered Users_State'] != merged['Registered Users_District']) |
    (merged['App Opens_State'] != merged['App Opens_District'])
]

# Display result
if discrepancies.empty:
    print(" No discrepancies found between state and district level data.")
else:
    print(" Discrepancies found between state and district level data:")
    print(discrepancies[['State', 'Year', 'Quarter']])
```

```python
# Task 4.1: Ratio of users to population by state


# Aggregate district-level population by state
state_population = df5.groupby('State')['Population'].sum().reset_index()

# Aggregate registered users at state level
state_users = df1.groupby('State')['Registered Users'].sum().reset_index()

# Merge population and users
state_pop_user = pd.merge(state_users, state_population, on='State')

# Calculate user-to-population ratio
state_pop_user['User to Population Ratio'] = state_pop_user['Registered Users'] / state_pop_user['Population']

# Display the result
print("User-to-Population Ratio by State:")
print(state_pop_user[['State', 'Registered Users', 'Population', 'User to Population Ratio']].sort_values(by='User to Population Ratio', ascending=False))
```

```python
# Task 4.4: Device brand usage ratio
# Merge district-level demographics with transactions
merged_district = pd.merge(
    df4,
    df5[['District', 'Code', 'Population', 'Density']],
    on=['District', 'Code'],
    how='inner'
)

# Aggregate transaction volume by district
txn_by_district = merged_district.groupby(['District', 'Code', 'Population', 'Density'])['Transactions'].sum().reset_index()

# Correlation between density and transaction volume
correlation = txn_by_district['Density'].corr(txn_by_district['Transactions'])
print(f"Correlation between population density and transaction volume: {correlation:.4f}")

# Scatter plot
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.scatter(txn_by_district['Density'], txn_by_district['Transactions'], alpha=0.6, color='darkorange')
plt.title('Population Density vs Transaction Volume')
plt.xlabel('Population Density')
plt.ylabel('Total Transactions')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Task 5.1: Plot total transactions and amount over time for a selected state

# Select state for visualization
selected_state = 'Maharashtra'

# Filter data for selected state
state_data = df1[df1['State'] == selected_state]

# Sort data by year and quarter
state_data = state_data.sort_values(by=['Year', 'Quarter'])

# Create time label for x-axis
state_data['Time'] = 'Q' + state_data['Quarter'].astype(str) + ' ' + state_data['Year'].astype(str)

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(state_data['Time'], state_data['Transactions'], marker='o', label='Transactions')
plt.plot(state_data['Time'], state_data['Amount (INR)'], marker='o', label='Transaction Amount (INR)')
plt.title(f'Transactions and Amount Over Time in {selected_state}')
plt.xlabel('Quarter')
plt.ylabel('Value')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Task 5.2: Pie chart of transaction types for a specific state and quarter

# Select state and quarter
selected_state = 'Maharashtra'
selected_year = 2021
selected_quarter = 1

# Filter transaction type data
txn_type_data = df2[
    (df2['State'] == selected_state) &
    (df2['Year'] == selected_year) &
    (df2['Quarter'] == selected_quarter)
]

# Group by transaction type and sum transactions
txn_grouped = txn_type_data.groupby('Transaction Type')['Transactions'].sum()

# Plot pie chart
plt.figure(figsize=(8, 8))
plt.pie(txn_grouped, labels=txn_grouped.index, autopct='%1.1f%%', startangle=140)
plt.title(f'Transaction Type Distribution in {selected_state} - Q{selected_quarter} {selected_year}')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
# Task 5.3: Visualize population density of districts in a selected state

# Select a state
selected_state = 'Tamil Nadu'

# Filter demographic data for the selected state
state_districts = df5[df5['State'] == selected_state]

# Sort by population density
state_districts = state_districts.sort_values(by='Density', ascending=False)

# Plot bar chart
plt.figure(figsize=(12, 6))
plt.bar(state_districts['District'], state_districts['Density'], color='salmon')
plt.title(f'Population Density of Districts in {selected_state}')
plt.xlabel('District')
plt.ylabel('Population Density')
plt.xticks(rotation=90)
plt.tight_layout()
plt.grid(axis='y')
plt.show()
```

```python
# Task 6.1: Identify trends or patterns in the transaction data

# Group data by year and quarter
overall_trends = df1.groupby(['Year', 'Quarter']).agg({
    'Transactions': 'sum',
    'Amount (INR)': 'sum'
}).reset_index()

# Create time labels
overall_trends['Time'] = 'Q' + overall_trends['Quarter'].astype(str) + ' ' + overall_trends['Year'].astype(str)

# Sort by time
overall_trends = overall_trends.sort_values(by=['Year', 'Quarter'])

# Plotting transaction trends
plt.figure(figsize=(12, 6))
plt.plot(overall_trends['Time'], overall_trends['Transactions'], marker='o', label='Total Transactions')
plt.plot(overall_trends['Time'], overall_trends['Amount (INR)'], marker='o', label='Total Amount (INR)')
plt.title('Overall Transaction Trends Over Time')
plt.xlabel('Quarter')
plt.ylabel('Value')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# Summary
print("\nTransaction Trends Summary:")
print("There is a steady increase in both the number of transactions and the total transaction amount over time, indicating growing adoption and usage of digital payments.")
```

```python
# Task 6.2: Correlate demographic data with transaction data

# Merge district transaction and demographic data
merged_df = pd.merge(df4, df5,
                     on=['State', 'District', 'Code'], how='inner')

# Compute correlation between population density and transaction volume
correlation = merged_df[['Density', 'Transactions']].corr().iloc[0, 1]
print(f"\nCorrelation between population density and transaction volume: {correlation:.2f}")

# Scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(merged_df['Density'], merged_df['Transactions'], alpha=0.6)
plt.title('Correlation between Population Density and Transactions')
plt.xlabel('Population Density')
plt.ylabel('Number of Transactions')
plt.grid(True)
plt.tight_layout()
plt.show()
```
